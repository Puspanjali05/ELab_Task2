1. Setting up the environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

2. Loading the dataset

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
df.head()
```

3. Basic data checks

```python
print(df.shape)
df.info()
df.describe(include='all')
df.isnull().sum()
```

4. Summary statistics

```python
print("Mean age:", df['Age'].mean())
print("Median age:", df['Age'].median())
print("Standard Deviation of Fare:", df['Fare'].std())
```

5. Visualizations

(i) Histograms

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

(ii) Boxplots

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Class')
plt.show()
```

(iii) Pairplot & Correlation

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

6. Identifying patterns or trends

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()

sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

7. Detect skewness and outliers

```python
print("Age Skewness:", df['Age'].skew())
print("Fare Skewness:", df['Fare'].skew())

sns.boxplot(df['Fare'])
plt.title("Outliers in Fare")
plt.show()
```
